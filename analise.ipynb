```python
# Renomear as colunas para facilitar o manuseio
data_f = data_f.rename(columns={
    'Ano': 'ano',
    'Código da Grande Região': 'cod_grande_regiao',
    'Nome da Grande Região': 'nome_grande_regiao',
    'Código da Unidade da Federação': 'cod_uf',
    'Sigla da Unidade da Federação': 'sigla_uf',
    'Nome da Unidade da Federação': 'nome_uf',
    'Código do Município': 'cod_municipio',
    'Nome do Município': 'nome_municipio',
    'Valor adicionado bruto da Agropecuária, \na preços correntes\n(R$ 1.000)': 'vab_agropecuaria',
    'Valor adicionado bruto da Indústria,\na preços correntes\n(R$ 1.000)': 'vab_industria',
    'Valor adicionado bruto dos Serviços,\na preços correntes \n- exceto Administração, defesa, educação e saúde públicas e seguridade social\n(R$ 1.000)': 'vab_servicos',
    'Valor adicionado bruto da Administração, defesa, educação e saúde públicas e seguridade social, \na preços correntes\n(R$ 1.000)': 'vab_adm_defesa_educacao_saude',
    'Valor adicionado bruto total, \na preços correntes\n(R$ 1.000)': 'vab_total',
    'Impostos, líquidos de subsídios, sobre produtos, \na preços correntes\n(R$ 1.000)': 'impostos_liquidos_subsidios',
    'Produto Interno Bruto, \na preços correntes\n(R$ 1.000)': 'pib_total',
    'Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)': 'pib_per_capita',
    'Atividade com maior valor adicionado bruto': 'atividade_maior_vab',
    'Atividade com segundo maior valor adicionado bruto': 'atividade_segundo_maior_vab',
    'Atividade com terceiro maior valor adicionado bruto': 'atividade_terceiro_maior_vab'
})



data_f.to_parquet("pib_municipios.parquet")
```

### Top 5 Ufs

```python
# Filtrando UFs por região (exemplo: Sudeste)
regiao_selecionada = 'Sudeste'
data_se = data_f[data_f['Nome da Grande Região'] == regiao_selecionada]

# Top 5 Estados por PIB em 2023
top_estados_2023 = data_se[data_se['Ano'] == 2023].groupby('Nome da Unidade da Federação')['Produto Interno Bruto, \na preços correntes\n(R$ 1.000)'].sum().nlargest(5)

# PIB ao longo dos anos  para os top 5 estados
top_estados_list = top_estados_2023.index.tolist()

pib_top_estados = data_se[(data_se['Nome da Unidade da Federação'].isin(top_estados_list))].groupby(['Ano', 'Nome da Unidade da Federação'])['Produto Interno Bruto, \na preços correntes\n(R$ 1.000)'].sum().unstack()

# Filtrar período de 2015 a 2023
pib_top_estados = pib_top_estados[(pib_top_estados.index >= 2015) & (pib_top_estados.index <= 2023)]

print(pib_top_estados)

```

### Valor adcionado por atividade econômica por todas as UFs 2015-2021

```python
### Filtrando por período 2015-2021

data_periodo = data_f[(data_f['Ano'] >= 2015) & (data_f['Ano'] <= 2021)]

# Tabela com porcentual do valor adicionado por atividade econômica por todas as UFs 2015-2021
tabela_atividades = data_periodo.groupby('Ano').agg({
    'Valor adicionado bruto da Agropecuária, \na preços correntes\n(R$ 1.000)': 'sum',
    'Valor adicionado bruto da Indústria,\na preços correntes\n(R$ 1.000)': 'sum',
    'Valor adicionado bruto dos Serviços,\na preços correntes \n- exceto Administração, defesa, educação e saúde públicas e seguridade social\n(R$ 1.000)': 'sum',
    'Valor adicionado bruto da Administração, defesa, educação e saúde públicas e seguridade social, \na preços correntes\n(R$ 1.000)': 'sum'
})
```

```python
# Ranking PIB per capita por UF região Nordeste 2023
data_ne = data_f[data_f['Nome da Grande Região'] == 'Nordeste']

# calculando
ranking_pib_per_capita_ne_2023 = data_ne[data_ne['Ano'] == 2023].groupby('Nome da Unidade da Federação')['Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)'].mean()

# Ranking PIB per capita por UF região Nordeste 2023
print(ranking_pib_per_capita_ne_2023.sort_values(ascending=False))
```

```python
# Considerando somente munícipios do Piauí
# Filtro para ajudar um scatter sobre relação do PIB com PIB per capita
data_pi = data_f[data_f['Nome da Unidade da Federação'] == 'Piauí']

### Dataframe final para análise do Piauí
data_pi_final = data_pi[['Ano', 'Nome do Município', 'Produto Interno Bruto, \na preços correntes\n(R$ 1.000)', 'Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)']]

# Somente ano 2023
data_pi_2023 = data_pi_final[data_pi_final['Ano'] == 2023]

print(data_pi_2023)
```

```python
# Gráfico consolidado por UF Sul, 
# UF, PIB, PIB per capita, Crescimento 2015-2021, Setor Dominante, Número de Municípios

# Primeiro, calcular crescimento do PIB 2015-2021 por UF Sul
data_sul = data_f[data_f['Nome da Grande Região'] == 'Sul']
pib_2015_sul = data_sul[data_sul['Ano'] == 2015].groupby('Nome da Unidade da Federação')['Produto Interno Bruto, \na preços correntes\n(R$ 1.000)'].sum()
pib_2021_sul = data_sul[data_sul['Ano'] == 2021].groupby('Nome da Unidade da Federação')['Produto Interno Bruto, \na preços correntes\n(R$ 1.000)'].sum()
crescimento_pib_sul = ((pib_2021_sul - pib_2015_sul) / pib_2015_sul) * 100

# Agora, gerar o gráfico com as informações solicitadas

data_sul_2021 = data_sul[data_sul['Ano'] == 2021].groupby('Nome da Unidade da Federação').agg({
    'Produto Interno Bruto, \na preços correntes\n(R$ 1.000)': 'sum',
    'Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)': 'mean',
    'Nome do Município': 'count'
})

data_sul_2021['Crescimento 2015-2021 (%)'] = crescimento_pib_sul
data_sul_2021 = data_sul_2021.rename(columns={'Nome do Município': 'Número de Municípios'})
data_sul_2021['Setor Dominante'] = data_sul_2021.index.map(lambda uf: data_sul[(data_sul['Nome da Unidade da Federação'] == uf) & (data_sul['Ano'] == 2021)].groupby('Nome do Município').agg({
    'Valor adicionado bruto da Agropecuária, \na preços correntes\n(R$ 1.000)': 'sum',
    'Valor adicionado bruto da Indústria,\na preços correntes\n(R$ 1.000)': 'sum',
    'Valor adicionado bruto dos Serviços,\na preços correntes \n- exceto Administração, defesa, educação e saúde públicas e seguridade social\n(R$ 1.000)': 'sum',
    'Valor adicionado bruto da Administração, defesa, educação e saúde públicas e seguridade social, \na preços correntes\n(R$ 1.000)': 'sum'
}).sum().idxmax())

print(data_sul_2021)
```
